```python
# Loading the dataset using pandas

import pandas as pd
csv_file_path_dir = r'C:/Users/tinat/OneDrive/Desktop/Customer behaviour Analysis/Dataset_Raw/archive/'
csv_file_name='shopping_data.csv'
csv_file_path=csv_file_path_dir + csv_file_name
df = pd.read_csv(csv_file_path)
df.head()
```

```python
df.info()
```

```python
# Summary statistics using .describe()
df.describe(include='all')
```

```python
#deleting un named coloumn
del df["Unnamed: 0"]
```

```python
df.info()
```

```python
# Checking if missing data or null values are present in the dataset

df.isnull().sum()
```

```python
# Inputing missing values in Review Rating column with the median rating of the product category

df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Renaming columns according to snake casing for better readability and documentation

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a new column age_group
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# Standardize purchase frequency labels:
# Replace "Fortnightly" with "Bi-Weekly"
# Replace "Every 3 Months" with "Quarterly"
df["frequency_of_purchases"] = df["frequency_of_purchases"].replace({
    "Fortnightly": "Bi-Weekly",
    "Every 3 Months": "Quarterly"
})
```

```python
df["frequency_of_purchases"].unique()
```

```python
# Create purchase tiers based on amount ranges
df["purchase_tier"] = pd.cut(
    df["purchase_amount"],
    bins=[0, 33, 66, 100],
    labels=["Low", "Medium", "High"]
)
```

```python
print(df["purchase_tier"].value_counts())
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# Dropping promo code used column

df = df.drop('promo_code_used', axis=1)
```

```python
# Install required libraries


from sqlalchemy import create_engine
from urllib.parse import quote_plus

server = "xxxxxxxxxxxxxx"   
database = "ShoppingData"

driver = quote_plus("ODBC Driver 18 for SQL Server")

engine = create_engine(
    f"mssql+pyodbc://@{server}/{database}?driver={driver}&trusted_connection=yes&TrustServerCertificate=yes"
)

# Test connection
with engine.connect() as conn:
    print("Connected successfully!")
```

```python


# --------------------------
# Write DataFrame to SQL Server
# --------------------------
# if_exists='replace' will overwrite the table
# index=False prevents adding DataFrame index as a column
df.to_sql("shopping_data", engine, if_exists="replace", index=False)

print("✅ Data written successfully to SQL Server!")
```
